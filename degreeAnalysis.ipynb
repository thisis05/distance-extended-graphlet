```python
import pandas as pd
import matplotlib.pyplot as plt

# txt 파일에서 데이터 읽기
file_path = './degree/ca-HepPh.txt'
with open(file_path, 'r') as file:
    header = file.readline().strip().split()  # 첫 번째 줄을 읽어 컬럼 이름으로 저장
    data = file.readlines()  # 나머지 데이터를 읽음

# 읽은 데이터로 데이터프레임 생성
df = pd.DataFrame([line.strip().split() for line in data], columns=header)

# 모든 값을 float로 변환
df = df.astype(float)

#df['outdeg2_times_indeg2'] = df['outdeg2'] * df['indeg2']

# 데이터 확인
print(df.head())
```

```python
# 그래프 그리기
plt.figure(figsize=(12, 6))

column_list = ['outdeg2', 'indeg2']
for col in column_list:
    plt.plot(df.index, df[col], label=col)

plt.xlabel('Node Index')
plt.ylabel('Degree')
plt.title('ca-HepPh : Node Degree Changes Over Index')
plt.legend()
#plt.yscale('log')
plt.grid(True)
plt.show()
```
